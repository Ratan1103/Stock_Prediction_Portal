```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Types of Plots 

## Line Plot

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.title('Simple Line Plot')
plt.xlabel('x-aixs')
plt.plot(x,y);
plt.grid(True)
```

## Bar Chart

```python
months=['Jan','Feb','Mar','Apr']
revenue=[100,300,200,500]
plt.figure(figsize=(10,6))
plt.bar(months,revenue);
plt.title('Revenue Bar Chart');
plt.ylabel('Revenue');
plt.xlabel('Months');
```

## Histogram

```python
marks=np.random.randint(0,100,50)
marks
```

```python
plt.hist(marks);
plt.xlabel('Marks');
plt.ylabel('No of Student');
```

## ScatterPlot 

```python
heights=[65,57,67,78,89]
weights=[150,160,165,176,177]
plt.scatter(heights,weights)
plt.title('Heights vs Weights Scatter Plot')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
```

## Subplots & Legends

```python
days=[1,2,3,4,5]
original_price=[10,20,25,30,40]
predicted_price=[8,21,25,28,38]
plt.plot(days,original_price,'b-o',label='Original Price')
plt.plot(days,predicted_price,'r-o',label='Predicted Price')
plt.legend()
plt.title('Plot with Legend')
plt.xlabel('Weekdays')
plt.ylabel('Price')
```

## Subplots

```python
plt.subplot(3,1,1)
plt.plot(days,original_price,'b-o',label='Original Price')
plt.title('Original Prices')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

plt.subplot(3,1,2)
plt.plot(days,predicted_price,'r-o',label='Predicted Price')
plt.title('Predicted Prices')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

diff=[(o-p) for o,p in zip(original_price,predicted_price)]
plt.subplot(3,1,3)
plt.bar(days,diff,color='purple',label='Difference Between Predicted and Original Price')
plt.xlabel('days')
plt.ylabel('Price Difference')
plt.tight_layout()
```

```python
days=[1,2,3,4,5]
original_price=[10,20,25,30,40]
predicted_price=[8,21,25,28,38]
plt.plot(days,original_price,'b-o',label='Original Price')
plt.plot(days,predicted_price,'r-o',label='Predicted Price')
plt.legend()
plt.title('Plot with Legend')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.savefig('example.png',dpi=300)
```
